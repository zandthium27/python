```python
Wichita = pd.read_csv("https://datajournalism.tech/wp-content/uploads/2019/10/wichita.csv") 
```

```python
Wichita.head()
```

```python
population = pd.DataFrame({"subject_race":["white", "black", "asian/pacific islander", "hispanic"], "n": [246343,42679,19272,63659]},columns=["subject_race","n"])
```

```python
population.head()
```

```python
population.info()
```

```python
four = Wichita [Wichita.subject_race != "other/unknown"]
```

```python
four.subject_race.value_counts()
```

```python
race = four.subject_race.value_counts().reset_index()
```

```python
race
```

```python
race = race.rename(columns = {"index": "subject_race", "subject_race":"count"})
```

```python
race
```

```python
population
```

```python
stops = pd.merge(race, population, on = "subject_race")
```

```python
stops
```

```python
stops["stop_rate"] = stops["count"]/stops["n"]*100
```

```python
stops
```

```python
stops = stops.sort_values("stop_rate", ascending=False)
```

```python
%matplotlib inline
```

```python
stops.stop_rate.plot.bar()
```

```python
stops.stop_rate.plot.barh()
```

```python
chart = stops.stop_rate.plot.barh()
chart.set_yticklabels(stops.subject_race)
```

```python
stops.to_csv("stops.csv")
```
